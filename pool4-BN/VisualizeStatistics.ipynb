```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def plot_statistics(stats):
    cells = []
    for i in (range(1, len(stats))):
        cells.append([])
        for j in range(len(stats[i])):
            cells[i-1].append(np.float64('%.4f' % np.float64(stats[i][j])))

    cells = np.array(cells)
    cells = cells[[3, 4, 1, 2, 0]]
    colors = cells * -0.6 + 1

    cells = np.transpose(cells)
    colors = np.transpose(colors)
    colors = np.array([colors] * 3)
    colors = np.transpose(colors, [1, 2, 0])

    rows = stats[0]
    cols = np.array(['True Match', 'False Match', 'AUROC', 'AUPR', 'BCE'])

    fig = plt.figure(figsize=(50, 25))
    fig.patch.set_visible(False)

    ax = fig.add_subplot()
    ax.axis('off')

    table = ax.table(cellText=cells, cellColours=colors, loc='center', rowLabels=rows, colLabels=cols)
    table.set_fontsize(25)
    table.scale(0.5, 4)

    plt.show()
    return fig
```

```python
categories = [i for i in os.walk('statistics')][0][2]
for i in range(len(categories)):
    print(categories[i].split('.')[0])
    plot_statistics(np.load(os.path.join('statistics', categories[i])))
```
